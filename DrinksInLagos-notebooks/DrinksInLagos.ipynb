```python
#Import tools

import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

from time import sleep
from random import randint
```

```python
#Get contents of the page we’re looking at by requesting the URL:

url = "https://www.drinksinlagos.com/product-category/wines/"

results = requests.get(url) 
```

```python
soup = BeautifulSoup(results.text, "html.parser")
```

```python
#find all of the lister-item mode-advanced divs:

wine_div = soup.find_all('div', class_='post_item_wrap')
```

```python
#initialize empty lists where we'll store our data
images = []
names = []
price = []
```

```python
for container in wine_div:
  
  #Name
  name = container.find("h2", class_ = "woocommerce-loop-product__title").text
  names.append(name)
        

  image = container.div.find("div", class_="post_thumb").img["src"]
  images.append(image)

  #price
  cost = container.find('span', class_='price').text 
  price.append(cost)
```

```python
DrinksInLagos_df = pd.DataFrame({"Name":names, "Images": images, "Price": price})
DrinksInLagos_df.head(50)
```
