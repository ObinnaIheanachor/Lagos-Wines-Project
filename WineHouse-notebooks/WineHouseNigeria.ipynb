```python
#Import tools

import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

from time import sleep
from random import randint
```

```python
pages = np.arange(1, 12, 1)
pages
```

```python
#initialize empty lists where we'll store our data
images = []
names = []
price = []
```

```python

for page in pages: 

  page = requests.get("https://winehousenigeria.com/product-category/wine/page/" + str(page) + "/")

  soup = BeautifulSoup(page.text, 'html.parser')
  wine_div = soup.find_all('li', class_= "product")
  
  sleep(randint(2,10))

  for container in wine_div:
      name = container.a.find("h2", class_="woocommerce-loop-product__title").text
      names.append(name)

      image = container.a.find("span", class_="et_shop_image").img["src"]
      images.append(image)

      cost = container.a.find("span", class_="woocommerce-Price-amount").text
      price.append(cost)

```

```python
print(len(names))
print(len(images))
print(len(price))
```

```python
wine_house_df = pd.DataFrame({"Name":names, "Images": images, "Price": price})
wine_house_df.head(60)
```

```python
wine_house_df["Name"] = wine_house_df["Name"].str.replace("\W", "")
```

```python
wine_house_df["Price"] = wine_house_df["Price"].str.replace("\W", "")
```

```python
wine_house_df.isnull().sum()
```

```python
wine_house_df.to_csv("WineHouseNigeria.csv", index=False)
```
